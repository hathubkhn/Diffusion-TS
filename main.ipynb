```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyts.image import GramianAngularField
from sklearn.preprocessing import MinMaxScaler

# Đọc dữ liệu CSV
df = pd.read_csv("SP500_values.csv")  # Thay "data.csv" bằng đường dẫn file của bạn

# Loại bỏ cột 'Symbol' và 'Date'
df_numeric = df.drop(columns=["Symbol", "Date"])

# Chuẩn hóa dữ liệu về [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))
data_normalized = scaler.fit_transform(df_numeric)

# Khởi tạo Gramian Angular Field
gaf = GramianAngularField(image_size=data_normalized.shape[1], method="summation")

# Tạo ảnh GAF cho từng cột
for i, column in enumerate(df_numeric.columns):
    gaf_image = gaf.fit_transform(data_normalized[:, i].reshape(1, -1))[0]
    
    # Hiển thị hình ảnh GAF
    plt.figure(figsize=(4, 4))
    plt.imshow(gaf_image, cmap="rainbow", origin="lower")
    plt.colorbar()
    plt.title(f"GAF - {column}")
    plt.savefig(f"gaf_{column}.png")  # Lưu ảnh
    plt.close()

print("Đã tạo ảnh GAF cho tất cả các cột!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyts.image import GramianAngularField
from sklearn.preprocessing import MinMaxScaler

# Đọc dữ liệu CSV
df = pd.read_csv("SP500_values.csv")

# Loại bỏ cột không cần thiết nếu có
df_numeric = df.drop(columns=["Symbol", "Date"], errors="ignore")

# Kiểm tra nếu dataframe rỗng
if df_numeric.empty:
    raise ValueError("Dữ liệu không hợp lệ! Không có cột số để xử lý.")

# Chuẩn hóa dữ liệu về [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))
data_normalized = scaler.fit_transform(df_numeric)

# Khởi tạo Gramian Angular Field
image_size = data_normalized.shape[0]  # Số bước thời gian làm kích thước ảnh
gaf = GramianAngularField(image_size=image_size, method="summation")

# Tạo ảnh GASF cho toàn bộ dữ liệu (đa biến)
gaf_image = gaf.fit_transform(data_normalized.T.reshape(1, data_normalized.shape[1], -1))[0]

# Hiển thị ảnh GASF
plt.figure(figsize=(6, 6))
plt.imshow(gaf_image, cmap="rainbow", origin="lower")
plt.colorbar()
plt.title("GASF - Multivariate Time Series")

# Lưu ảnh dưới dạng PNG với cấu hình chất lượng cao
plt.savefig("gaf_multivariate.png", dpi=300, bbox_inches="tight")

# Hiển thị ảnh
plt.show()

print("Đã tạo ảnh GASF cho toàn bộ chuỗi thời gian đa biến!")
```

```python
import numpy as np
import pandas as pd
from pyts.image import GramianAngularField
from sklearn.preprocessing import MinMaxScaler
from PIL import Image

# Đọc dữ liệu CSV (giới hạn số dòng để tránh quá tải bộ nhớ)
df = pd.read_csv("SP500_values.csv")

# Loại bỏ cột không cần thiết nếu có
df_numeric = df.drop(columns=["Symbol", "Date"], errors="ignore")

# Kiểm tra nếu dataframe rỗng
if df_numeric.empty:
    raise ValueError("Dữ liệu không hợp lệ! Không có cột số để xử lý.")

# Chuẩn hóa dữ liệu về [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))
data_normalized = scaler.fit_transform(df_numeric)

# Chọn một biến (cột) để tạo ảnh GASF (nếu có nhiều cột, chọn cột đầu tiên)
selected_feature = data_normalized[:, 0].reshape(1, -1)  # Dữ liệu 2D dạng (1, n_timestamps)

# Giảm kích thước ảnh để tránh lỗi MemoryError
image_size = min(data_normalized.shape[0], 100)  # Giới hạn tối đa 100x100 pixel
gaf = GramianAngularField(image_size=image_size, method="summation")

# Chuyển đổi dữ liệu thành ảnh GASF
gaf_image = gaf.fit_transform(selected_feature)[0]

# Chuẩn hóa giá trị ảnh về dạng uint8 (0-255)
gaf_image_normalized = ((gaf_image - gaf_image.min()) / (gaf_image.max() - gaf_image.min()) * 255).astype(np.uint8)

# Chuyển thành ảnh PIL và lưu file
image = Image.fromarray(gaf_image_normalized)
image.save("gaf_image.png")

print("Ảnh GASF đã được lưu thành công dưới dạng PNG!")
```

```python
import pandas as pd
import numpy as np
import os
from pyts.image import GramianAngularField
from sklearn.preprocessing import MinMaxScaler
from PIL import Image

seq_len = 96
input_file = 'SP500_values.csv' 
output_dir = 'gasf_png_rgb'
os.makedirs(output_dir, exist_ok=True)
df = pd.read_csv(input_file, usecols=['Symbol', 'Date', 'Close'])
df['Date'] = pd.to_datetime(df['Date'])

for symbol in df['Symbol'].unique():
    df_symbol = df[df['Symbol'] == symbol].sort_values('Date')
    close_prices = df_symbol['Close'].values

    # scale
    scaler = MinMaxScaler(feature_range=(-1, 1))
    normalized = scaler.fit_transform(close_prices.reshape(-1, 1)).flatten()

    # GASF
    gasf = GramianAngularField(image_size=seq_len, method='summation')
    symbol_dir = os.path.join(output_dir, symbol)
    os.makedirs(symbol_dir, exist_ok=True)

    for i in range(len(normalized) - seq_len + 1):
        window = normalized[i:i + seq_len].reshape(1, -1)
        gasf_image = gasf.fit_transform(window)[0]

        gasf_scaled = ((gasf_image - np.min(gasf_image)) / (np.max(gasf_image) - np.min(gasf_image)) * 255).astype(np.uint8)

        
        rgb_image = np.stack([gasf_scaled]*3, axis=-1)

        img = Image.fromarray(rgb_image, mode='RGB')
        date_str = df_symbol.iloc[i + seq_len - 1]['Date'].strftime('%Y%m%d')
        img.save(os.path.join(symbol_dir, f"{date_str}.png"))

    print(f"✅ Đã xử lý xong symbol: {symbol}")
```
